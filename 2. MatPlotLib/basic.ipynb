# **MatPlotLib**
### Python's most popular 2-D plotting library
- A range of available graph types like bar, line, histogram, etc.
- Allows to plot from various data types like lists, numpy arras, pandas dataframe, etc.
- Gives full control of other suppliments like colour, styles, font, axes, etc.

<hr>

Importing Library

```python
import matplotlib.pyplot as plt
```

A Simple Plot

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.title('Squares of numbers', fontsize=24, fontname='Comic Sans MS')
plt.ylabel('Squares', color='red')
plt.xlabel('Numbers', color='red')
plt.show()
```

Customising the plots

```python
plt.figure(figsize=(10, 5))                                     # in inches
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'rx')                     # red x
plt.axis([0, 6, 0, 20])                                         # [xmin, xmax, ymin, ymax]
plt.show()
```

Annotations

```python
plt.xticks([1, 2, 3, 4], ['A', 'B', 'C', 'D'])
plt.bar([1, 2, 3, 4], [1, 4, 9, 16], color=['yellow', 'blue', 'green', 'red'])
plt.annotate('AviLOL', xy=(2, 8), xytext=(3, 15), arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

Third-party libraries

```python
import numpy as np
import pandas as pd

ts = pd.Series(np.random.randn(10), index=pd.date_range('1/1/2000', periods=10))
df = pd.DataFrame(np.random.randn(10, 4), index=ts.index, columns=list('ABCD'))
df.plot(style=['ro-', 'b^-', 'gs--', 'cv-.'])

plt.show()
```

Scatter

```python
import numpy as np
import matplotlib.pyplot as plt

N = np.random.randint(69)
x = np.random.randn(N)
y = np.random.randn(N)
colour = np.random.rand(N)
size = np.random.rand(N) * 370

plt.scatter(x, y, s=size, c=colour, alpha=0.5)
plt.show()
```

Subplots

```python
fig = plt.figure()
fig.suptitle('ur mom')

fig.add_subplot(221, title='1')
fig.add_subplot(222)
plt.plot([5, 6, 7, 8], [25, 36, 49, 64])
fig.add_subplot(223)
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
fig.add_subplot(224, facecolor='b')

plt.show()
```

Another way of subplotting

```python
import numpy as np

fig, plots = plt.subplots(2, sharex=True)
fig.suptitle('Sharing X axis')

x, y = np.arange(5), np.arange(5)
plots[0].plot(x, y)
plots[1].scatter(x, y)

plt.show()
```

Saving a plot as an image file

```python
plt.stem([1, 2, 3, 4], [1, 4, 9, 16])
plt.savefig('plot.png')
```

Clearing a plot

```python

plt.clf()
```
